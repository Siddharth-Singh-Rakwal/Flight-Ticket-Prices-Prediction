```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")

%matplotlib inline
```

```python
# Importing Dataset

df = pd.read_excel('flight_price.xlsx')
df.head()
```

```python
# Check for Null Value in Dataset:

df.isnull().sum()
```

```python
# Dropping Null Value in Dataset:

df.dropna(inplace = True)
```

```python
df.isnull().sum()
```

```python
# Extracting 'Date_of_Journey' Column into 'Journey_Month' & 'Journey_Date' Column:

df["Journey_Date"] = pd.to_datetime(df.Date_of_Journey, format="%d/%m/%Y").dt.day
```

```python
df["Journey_Month"] = pd.to_datetime(df["Date_of_Journey"], format = "%d/%m/%Y").dt.month
```

```python
df.head(1)
```

```python
# Dropping 'Date_of_Journey' Column:

df.drop(["Date_of_Journey"], axis = 1, inplace = True)
```

```python
df.head(1)
```

```python
# Extracting 'Dep_Time' Column:

# Extracting Hours:
df["Dep_Hour"] = pd.to_datetime(df["Dep_Time"]).dt.hour

# Extracting Minutes:
df["Dep_Minutes"] = pd.to_datetime(df["Dep_Time"]).dt.minute
```

```python
df.head(1)
```

```python
# Dropping 'Dep_Time' Column:

df.drop(["Dep_Time"], axis = 1, inplace = True)

df.head(1)
```

```python
# Extracting 'Arrival_Time' Column:

# Extracting Hours:
df['Arrival_Hour'] = pd.to_datetime(df.Arrival_Time).dt.hour

# Extracting Minutes:
df['Arrival_Minutes'] = pd.to_datetime(df.Arrival_Time).dt.minute
```

```python
df.head(1)
```

```python
# Dropping 'Arrival_Time' Column:

df.drop(["Arrival_Time"], axis=1, inplace=True)

df.head(1)
```

```python
# Converting 'Duration' Column into List:

duration = list(df["Duration"])

for i in range(len(duration)):
    if len(duration[i].split()) != 2:     # Check if Duration contains only hours or mins
        if "h" in duration[i]:
            duration[i] = duration[i].strip() + " 0m"  # Add 0 Minutes
        else:
            duration[i] = "0h " + duration[i]     # Add 0 Hour
            
duration_hours = []
duration_mins = []

for i in range(len(duration)):
    duration_hours.append(int(duration[i].split(sep="h")[0]))  # Extract hours from Duration
    duration_mins.append(int(duration[i].split(sep="m")[0].split()[-1]))    # Extract Minutes from Duration
```

```python
# Adding Duration_hours and Duration_mins Column to our Dataset:

df["Duration_Hours"] = duration_hours
df["Duration_Minutes"] = duration_mins
```

```python
df.head(1)
```

```python
# Dropping 'Duration' Column:

df.drop(["Duration"], axis=1, inplace=True)
```

```python
df.head(1)
```

```python
# Does Price varies with Airlines?

plt.figure(figsize =(8,6))
sns.barplot(x=df['Airline'], y= df['Price'],palette='hls')
plt.title("Airlines Vs Price", fontsize=15)
plt.xlabel("Airline", fontsize=15)
plt.ylabel("Price", fontsize=15)
plt.xticks(rotation = 90)
plt.show()
```

```python
df["Airline"].value_counts()
```

```python
# Handling Outlires:
# Price vs Airline:

sns.catplot(y = "Price", x = "Airline", data = df.sort_values("Price", ascending = False), kind="boxen", height = 6, aspect = 3)
plt.show()
```

```python
# As Airline is Nominal Categorical data we will perform OneHotEncoding:

Airline = df[["Airline"]]

Airline = pd.get_dummies(Airline, drop_first= True)

Airline.head()
```

```python
df["Source"].value_counts()
```

```python
# Source vs Price:

sns.catplot(y = "Price", x = "Source", data = df.sort_values("Price", ascending = False), kind="boxen", height = 4, aspect = 3)
plt.show()
```

```python
# As Source is Nominal Categorical data we will perform OneHotEncoding

Source = df[["Source"]]

Source = pd.get_dummies(Source, drop_first= True)

Source.head()
```

```python
df["Destination"].value_counts()
```

```python
# As Destination is Nominal Categorical data we will perform OneHotEncoding

Destination = df[["Destination"]]

Destination = pd.get_dummies(Destination, drop_first = True)

Destination.head()
```

```python
df["Route"]
```

```python
# Additional_Info contains almost 80% no info
# Route and Total_Stops are related to each other

df.drop(["Route", "Additional_Info"], axis = 1, inplace = True)
```

```python
df["Total_Stops"].value_counts()
```

```python
# As this is case of Ordinal Categorical type, we perform LabelEncoder
# Here Values are assigned with corresponding keys

df.replace({"non-stop": 0, "1 stop": 1, "2 stops": 2, "3 stops": 3, "4 stops": 4}, inplace = True)
```

```python
df.head(2)
```

```python
# Concatenate dataframe --> train_data + Airline + Source + Destination:

df = pd.concat([df, Airline, Source, Destination], axis = 1)
df.head()
```

```python
# Dropping 'Airline', 'Source', 'Destination' Column:

df.drop(["Airline", "Source", "Destination"], axis = 1, inplace = True)
```

```python
df.head(1)
```

```python
# Save the Cleaned Dataset:

df.to_csv('Flight_Cleaned_Dataset.csv',index=False)
```
